```python
# 1.a) I found some data about the popularity of different noodle dishes. I downloaded it as trends.csv file:

trends_path = "C:/Users/Nik/Documents/trends.csv"
trends_df = pd.read_csv(trends_path)
trends_df.head(2)
```

```python
# Slicing out the data I want for my timeseries: Number of Google searches about Ramen per month since 2004:

trends_by_month = trends_df.set_index("Month")
ramen_trends = pd.DataFrame(trends_by_month['Ramen'])
ramen_trends
```

## 1.b&c) For my data, the most logical choice for timeseries analysis is ARIMA. The data is not stationary, so it would be best to find the order of integration.

```python
plt.plot(ramen_trends.index, ramen_trends['Ramen'])
```

```python
#The above data does not look stationary. I will transform to a stationary process
# to see if the data makes sense that way.
ramen_trends['Stationary']=ramen_trends['Ramen'].pct_change()
ramen_trends['Stationary'].plot()
```

```python
#autocorrelation used to estimate MA order (5)
plot_acf(ramen_trends[['Ramen']], lags=5, missing='drop')
```

```python
#PCAF used to estimate AR order (1)
plot_pacf(ramen_trends[['Ramen']], zero=False, lags = 10)
```

```python
# 1.c) Run either an ARMA, ARIMA, or SARIMA on it
# I ran an ARIMA model because there is a noticeable 
#improvement to data clarity after converting to stationary process
# but data is not seasonal. For my model I used lag values of of AR = 1, I=1, and MA=5.


model = ARIMA(ramen_trends['Ramen'], order=(1, 1, 5))

#fit the model to the data
results = model.fit()

#create a dataframe
results_df = pd.DataFrame(results.forecast(steps=10))
results_df.plot(title='Ramen')
plt.show()
results.summary()
```

2. What is a fourier transform? Perform one on y = np.cos(x/2), and both show (with code) and explain the key elements

```python

x = np.arange(1,50,1)
y = np.cos(x/2)

#Let's plot the given cosine wave before we transform it:

plt.plot(x,y,'.-')
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('Cosine Wave')
plt.show()
```

```python
#here is the numpy function for the fast fourier transformation:
fft_cos=np.fft.rfft(y)

# Power spectral densities (PSD) are used to characterize random vibration signals. The key aspect of a PSD is the
# amplitude value normalized to the frequency bin width. By normalizing the amplitude we get rid of the dependency so that we
# can compare vibration levels in signals of different lengths.

#calculating the power spectral density
spect_density = np.abs(fft_cos)
frequency = np.fft.rfftfreq(len(y))

#plotting FFT of our input signal
plt.plot(frequency, spect_density)
#Finding peak value
plt.axvline(1/(4*np.pi),c='r')
plt.xlabel("Frequency")
plt.ylabel("Power Spectral Density")
plt.show()
```
